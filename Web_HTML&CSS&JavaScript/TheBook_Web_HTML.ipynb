## 웹 사이트 레이아웃

-	정적 레이아웃
  1. 고정 된 크기, 너비를 줄이면 가로 스크롤이 나옴
  2. 스마트폰으로 보면 스크롤이 생겨 불편
-	동적 레이아웃
  1.	크기 상관없이 화면에 콘텐츠가 꽉 채워짐
  2.	글귀 나열 방식이 달라질 수 있음
  3.	스크롤 없음
-	적응형 레이아웃
  1. 정적 레이아웃 처럼 크기는 고정 + 사용자가 정의한 여러 너비에 따라 레이아웃 형태가 바뀜
  2. 스마트폰 , 태블릿 등 다양한 기기 화면 지원
  3. 반응형 레이아웃보다 제작 시간 적게 듬
-	반응형 레이아웃
  1. 동적 레이아웃 (브라우저 너비에 따라 크기 달라짐) + 적응형 레이아웃 (공간의 배치가 달라짐) 결합
